```python
#Realiza la distribucion de tablas
df_customers = df_store.iloc[:,5:13] #tabla customer usando slicing
df_products = df_store[['Product ID','Product Name','Category','Sub-Category']] #tabla products
df_sales = df_store[['Row ID','Order ID','Order Date','Ship Date','Ship Mode','Customer ID','Product ID','Sales','Quantity','Discount','Profit']].copy() #tabla sales
```

```python
#Quitando valores duplicados
df_customers = df_customers[['Customer ID', 'Customer Name', 'Segment', 'Country', 'City', 'State', 'Postal Code', 'Region']].drop_duplicates(subset='Customer ID')
df_products = df_products[['Product ID', 'Product Name', 'Category', 'Sub-Category']].drop_duplicates(subset='Product ID')
```

```python
# Cantidad de duplicados en la clave primaria
print(df_customers['Customer ID'].duplicated().sum()) 
print(df_products['Product ID'].duplicated().sum())   
print(df_sales['Order ID'].duplicated().sum())   
```
